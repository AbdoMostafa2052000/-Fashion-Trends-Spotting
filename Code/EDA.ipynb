## Brand

```python
# Count plot of the number of posts per brand
plt.figure(figsize=(14, 8))
sns.countplot(data=df, x='Brand', order=df['Brand'].value_counts().index)
plt.title('Number of Posts per Brand')
plt.xlabel('Brand')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

## Trend Analysis Over Time for Post Likes

```python
# Trend analysis over time for post likes
plt.figure(figsize=(14, 8))
df.groupby(df['post_date'].dt.to_period("M"))['post_likes'].sum().plot(kind='bar')
plt.title('Total Post Likes Over Time')
plt.xlabel('Date')
plt.ylabel('Total Post Likes')
plt.xticks(rotation=45)
plt.show()
```
