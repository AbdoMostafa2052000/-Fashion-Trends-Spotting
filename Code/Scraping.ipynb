```python
import instaloader
from IPython.display import display
import pandas as pd
from instaloader.exceptions import BadResponseException, ConnectionException, LoginRequiredException, TooManyRequestsException
```

```python
il = instaloader.Instaloader()
```

```python
# Initialize Instaloader
L = instaloader.Instaloader()
#Function to login to Instagram
def login_instagram(username, password):
    try:
        L.login(username, password)
        print("Logged in to Instagram")
    except (BadResponseException, ConnectionException, LoginRequiredException) as e:
        print(f"Error during login: {e}")

# Function to fetch post data from a profile
def fetch_profile_posts_data(profile_name):
    try:
        # Load the profile
        profile = instaloader.Profile.from_username(L.context, profile_name)

        # Initialize lists to hold the post data
        data = {
            "post_id": [],
            "post_shortcode": [],
            "post_date": [],
            "post_caption": [],
            "post_likes": [],
            "image_url": [],

            #"title":[],
            #"typename":[],

            "post_is_video": [],
            #"post_video_views": [],
            "post_hashtags": [],
            "caption_hashtags":[],
            "post_mentions": [],
            "post_tagged_users": [],


            "commenter_username": [],
            "comment_text": [],
            "comment_date": [],
            "comment_no_likes": []

        }
        i = 0
        # Iterate over posts in the profile
        for post in profile.get_posts():
            if i == 200:
                break
            # Iterate over comments
            j = 0
            number_comm=len(list(post.get_comments()))
            for comment in post.get_comments():
                if j == max(number_comm // 20, 100):
                    break
                data["post_id"].append(post.mediaid)
                data["post_shortcode"].append(post.shortcode)
                data["post_date"].append(post.date)
                data["post_caption"].append(post.caption)
                data["post_likes"].append(post.likes)
                data["image_url"].append(post.url)

                # data["title"].append(post.title)
                # data["typename"].append(post.typename)


                data["post_is_video"].append(post.is_video)
                #data["post_video_views"].append(post.video_view_count if post.is_video else None)
                data["post_hashtags"].append(post.caption_hashtags)
                data["caption_hashtags"].append(post.caption_hashtags)
                data["post_mentions"].append(post.caption_mentions)
                data["post_tagged_users"].append([user for user in post.tagged_users])

                data["commenter_username"].append(comment.owner.username)
                data["comment_text"].append(comment.text)
                data["comment_date"].append(comment.created_at_utc)
                data["comment_no_likes"].append(comment.likes_count)
                j+=1
            print(i)
            i+=1
        # Convert the data to a DataFrame
        df = pd.DataFrame.from_dict(data , orient='index').T
        return df

    except (BadResponseException, ConnectionException, TooManyRequestsException) as e:
        print(f"Error fetching profile data: {e}")
        return None

login_instagram(username, password)
df = fetch_profile_posts_data(profile_name)

# Display the DataFrame if fetched successfullynt
if df is not None:
    #print(df)
    # Optionally, save the DataFrame to a CSV file
    df.to_csv(profile_name+'.csv', index=False)
filename = profile_name+'.csv'
destination_path = "/content/drive/MyDrive/"
!cp {filename} {destination_path}
```
